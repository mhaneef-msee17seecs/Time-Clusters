```python
#Dimensionality Reduction - Low Variance Filter
features= traindata[['time[ps]', 'address', 'binaryaddress', 'bank', 'row', 'column', 'command [XXX]']]
var = features.var()

features = features.columns
variable = [ ]
for i in range(0,len(var)):
    if var[i]>=10:   #setting the threshold as 10%
       variable.append(features[i+1])
    
print(variable)
```

```python
#Dimensionality Reduction - Random Forest Regressor
df = traindata.drop('delay[ps]', 1)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1, max_depth=10)
df=pd.get_dummies(df)
model.fit(df, traindata['delay[ps]'])
```

```python
import numpy as np
features = df.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-9:]  # top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

traindata = pd.read_csv('train_debrujin.txt', sep = ',')
testdata = pd.read_csv('test_filter.txt', sep=',')

features= ['time[ps]', 'address', 'binaryaddress', 'bank', 'row', 'column', 'command [XXX]']

traindata = traindata.rename(columns={' binaryaddress': 'binaryaddress'})
traindata = traindata.rename(columns={' bank': 'bank'})
traindata = traindata.rename(columns={' row': 'row'})
traindata = traindata.rename(columns={' column': 'column'})
traindata = traindata.rename(columns={' delay [ps]': 'delay[ps]'})
traindata = traindata.rename(columns={' command [XXX]': 'command [XXX]'})
traindata = traindata.rename(columns={'time [ps]': 'time[ps]'})
traindata = traindata.rename(columns={' address':'address'})

testdata = testdata.rename(columns={' binaryaddress': 'binaryaddress'})
testdata = testdata.rename(columns={' bank': 'bank'})
testdata = testdata.rename(columns={' row': 'row'})
testdata = testdata.rename(columns={' column': 'column'})
testdata = testdata.rename(columns={' delay [ps]': 'delay[ps]'})
testdata = testdata.rename(columns={' command [XXX]': 'command [XXX]'})
testdata = testdata.rename(columns={'time [ps]': 'time[ps]'})
testdata = testdata.rename(columns={' address':'address'})

X_train = traindata[features]
y_train = traindata['delay[ps]']

X_test = testdata[features]
y_test = testdata['delay[ps]']

clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
confidence = clf.score(X_test, y_test)
print(confidence)
```
